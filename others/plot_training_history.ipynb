```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(palette='Paired')
```

```python
df = pd.read_csv('training_history_by_csv_logger.csv')
```

```python
from functools import reduce
flatten =lambda l: reduce(lambda x, y: x + y, l) 
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, dpi=120, figsize=(20, 30))
df.plot(y=['loss'], legend=False, ax=ax1)
ax1.legend(loc='best', fontsize=8); ax1.set_title('Training loss')
df.plot(y=['val_loss'], legend=False, ax=ax2)
ax2.legend(loc='best', fontsize=8); ax2.set_title('Testing loss')
df.plot(y=['output_{}_acc'.format(i+1) for i in range(6)], legend=False, ax=ax3)
ax3.legend(loc='best', fontsize=8); ax3.set_ylim(0, 1); ax3.set_title('Training acc')
df.plot(y=['val_output_{}_acc'.format(i+1) for i in range(6)], legend=False, ax=ax4)
ax4.legend(loc='best', fontsize=8); ax4.set_ylim(0, 1); ax4.set_title('Testing acc')
df.plot(y=['output_{}_loss'.format(i+1) for i in range(6)], legend=False, ax=ax5)
ax5.legend(loc='best', fontsize=8); ax5.set_title('Training loss'); ax5.set_title('Training loss')
df.plot(y=['val_output_{}_loss'.format(i+1) for i in range(6)], legend=False, ax=ax6)
ax6.legend(loc='best', fontsize=8); ax6.set_title('Testing loss')
#ax1.legend('upper left', ncol=4)
```
