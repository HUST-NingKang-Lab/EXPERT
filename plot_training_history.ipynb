```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(palette='Paired')
```

```python
df = pd.read_csv('training_history.csv')
```

```python
from functools import reduce
flatten =lambda l: reduce(lambda x, y: x + y, l) 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, dpi=120, figsize=(10, 10))
df.plot(y=['l{}_acc'.format(i) for i in range(6)], legend=False, ax=ax1)
ax1.legend(loc='best', fontsize=8); ax1.set_ylim(0, 1)
df.plot(y=['val_l{}_acc'.format(i) for i in range(6)], legend=False, ax=ax2)
ax2.legend(loc='best', fontsize=8); ax2.set_ylim(0, 1)
df.plot(y=['l{}_loss'.format(i) for i in range(6)], legend=False, ax=ax3)
ax3.legend(loc='best', fontsize=8); ax3.set_ylim(0, 1)
df.plot(y=['val_l{}_loss'.format(i) for i in range(6)], legend=False, ax=ax4)
ax4.legend(loc='best', fontsize=8); ax4.set_ylim(0, 1)
#ax1.legend('upper left', ncol=4)
```
